```python
import google.generativeai as genai
import os
```

```python
from google.colab import userdata
GEMINI_API_KEY=userdata.get('GEMINI_API_KEY2')
```

```python
# 2. Setup - REPLACE 'GEMINI_API_KEY' WITH YOUR ACTUAL KEY
os.environ["GOOGLE_API_KEY"] = "GEMINI_API_KEY"
genai.configure(api_key=GEMINI_API_KEY)
```

```python
# 3. Initialize the Model
model = genai.GenerativeModel('gemini-2.5-flash')
```

```python
print("--- Interaction 1: Teaching the model a fact ---")
# We tell the model a specific fact (a secret number)
prompt_1 = "Hi! Remember this secret number: 998877. Do not forget it."
response_1 = model.generate_content(prompt_1)
print(f"Me: {prompt_1}")
print(f"Gemini: {response_1.text}")
```

```python
print("\n--- Interaction 2: Testing Memory ---")
# We ask the model to recall the fact in a NEW call
prompt_2 = "What is the secret number I just told you?"
response_2 = model.generate_content(prompt_2)
print(f"Me: {prompt_2}")
print(f"Gemini: {response_2.text}")
```

Notice that in Interaction 2, Gemini likely said it doesn't know or asked for context.
This proves the model itself has no memory; it treats every call as a brand new conversation.
