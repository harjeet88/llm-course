```python
import tiktoken

tokenizer = tiktoken.get_encoding("cl100k_base")
```

```python
text = "I love unbelievable transformations!"
```

```python
tokens = tokenizer.encode(text)
decoded_tokens = [tokenizer.decode([t]) for t in tokens]

print("Tokens:", tokens)
print("Decoded pieces:", decoded_tokens)
```
